```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme()
```

```python
df_reddit = pd.read_csv("data/sentiment-analysis/reddit-sentiment-analysis.csv")
df_twitter = pd.read_csv("data/sentiment-analysis/twitter-sentiment-analysis.csv")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
sns.countplot(data=df_reddit, x="category", ax=ax[0])
ax[0].set_xlabel("Sentiment")
ax[0].set_ylabel("Count")
ax[0].set_title("Sentiment Distribution in Reddit Dataset")

sns.countplot(data=df_twitter, x="category", ax=ax[1])
ax[1].set_xlabel("Sentiment")
ax[1].set_ylabel("Count")
ax[1].set_title("Sentiment Distribution in Twitter Dataset")
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
def get_sentiment_score(text):
    try:
        return sid.polarity_scores(text)["compound"]
    except:
        return 0

get_sentiment_score("yay...")
```
