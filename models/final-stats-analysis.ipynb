```python
with open(f"./final/all-y-hat-ns.pkl", "rb") as f:
    y_hat_ns = pkl.load(f)

with open(f"./final/all-y-hat-s.pkl", "rb") as f:
    y_hat_s = pkl.load(f)

with open(f"./final/all-y-test-ns.pkl", "rb") as f:
    y_test_ns = pkl.load(f)

with open(f"./final/all-y-test-s.pkl", "rb") as f:
    y_test_s = pkl.load(f)

y_test_ns = [x.reshape(-1) for x in y_test_ns]
y_test_s = [x.reshape(-1) for x in y_test_s]
```

```python
for a, b in zip(y_hat_ns, y_hat_s):
    print(a.shape, b.shape)
```

```python
ns_15 = y_hat_ns[0]
ns_30 = y_hat_ns[1]
ns_60 = y_hat_ns[2]
ns_90 = y_hat_ns[3]

s_15 = y_hat_s[0]
s_30 = y_hat_s[1]
s_60 = y_hat_s[2]
s_90 = y_hat_s[3]

y_test_15 = y_test_ns[0]
y_test_30 = y_test_ns[1]
y_test_60 = y_test_ns[2]
y_test_90 = y_test_ns[3]
```

```python
# rmse between y_test and y_hat (15)
stock_tickers = ["META", "GOOGL", "NFLX", "AMZN", "TSLA", "HD", "PG", "COST", "PEP", "XOM", "CVX", "COP", "MA", "JPM", "V", "UNH", "JNJ", "LLY", "UNP", "BA", "CAT", "LIN", "SHW", "FCX", "PLD", "AMT", "EQIX", "MSFT", "AAPL", "AVGO", "NEE", "SO", "DUK"]

df_ticker_rmse = pd.DataFrame(columns=["Ticker", "Sentimentality", "Prediction", "Truth"])

for s in stock_tickers:
    sentimentality = ["Sentiment"] * len(y_test_15[test_tickers_15 == s])
    sentimentality.extend(["No Sentiment"] * len(y_test_15[test_tickers_15 == s]))
    df = pd.DataFrame({
        "Ticker": [s] * 2 * len(y_test_15[test_tickers_15 == s]),
        "Sentimentality": sentimentality,
        "Prediction": np.concatenate([s_15[test_tickers_15 == s], ns_15[test_tickers_15 == s]]),
        "Truth": np.concatenate([y_test_15[test_tickers_15 == s], y_test_15[test_tickers_15 == s]])
    })
    
    df_ticker_rmse = pd.concat([df_ticker_rmse, df], ignore_index=True)

df_ticker_rmse["Square Error"] = (df_ticker_rmse["Prediction"] - df_ticker_rmse["Truth"]) ** 2
df_ticker_rmse
```
